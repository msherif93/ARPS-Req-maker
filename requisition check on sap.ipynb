```python
import pandas as pd
import numpy as np
```

```python
sap=pd.read_csv('2042.csv',encoding='ISO-8859-1')
```

```python
req=pd.read_csv('OFFER.csv',encoding='ISO-8859-1')
```

```python
req.head(3)
```

```python
req1=req[['PTB CODE','O&G Material Description','OFFER Qty']]
```

```python
req1=req1[~req1['O&G Material Description'].isnull()]
```

```python
req1['PTB CODE']=req1['PTB CODE'].str.replace('\n','')
```

```python
req1
```

```python
sap.info()
```

```python
sap1=sap[['Plant','Storage Location','Material Code','Unrestricted','Batch']]
sap1=sap1[~sap1['Storage Location'].isnull()]
sap1['Unrestricted']=sap1['Unrestricted'].str.replace(',','',regex=False).astype(float)
```

```python
req1=req1[req1['PTB CODE']!='N.C']
```

```python
p=[]
for code in req1['PTB CODE']:
    temp=sap1[sap1['Material Code']==int(code)]
    plants=temp.Plant.unique()
    p.extend(plants)
    
plants=list(set(p))
for plant in plants:
    req1[plant]=0
req1.head()
```

```python
for code in req1['PTB CODE']:
    temp=sap1[sap1['Material Code']==int(code)]
    plants=temp.Plant.unique()

    for plant in plants:

            x=temp[temp['Plant']==plant]
            index=req1[req1['PTB CODE']==code][plant].index[0]
            req1.iloc[index,req1.columns.get_loc(plant)]=x['Unrestricted'].sum()
            #print(req1[req1['PTB CODE']==code][plant])
            #print(x['Unrestricted'].sum())
            
```

```python
req1.head()
```

```python
covered_material=req1[req1['ARPE']>=req1['OFFER Qty']]
```

```python
covered=covered_material['PTB CODE'].tolist()
not_covered=req1[~req1['PTB CODE'].isin(covered)]
```

```python
not_covered['ACTUALNEED']=not_covered['OFFER Qty']-not_covered['ARPE']
```

```python

not_covered['Total other fields']=not_covered[list(set(p))[1:]].sum(axis=1)
```

```python
direct_to_req=not_covered[not_covered['Total other fields']==0]
direct_to_req
```

```python
other_fields=not_covered[~not_covered['PTB CODE'].isin(direct_to_req['PTB CODE'].tolist())]
```

```python
other_fields['PF_TOTAL']=other_fields[total_pf].sum(axis=1)
other_fields['AM_TOTAL']=other_fields[total_abumadi].sum(axis=1)
```

```python
other_fields['PF_required']=np.where(other_fields['ACTUALNEED']<other_fields['PF_TOTAL'],other_fields['ACTUALNEED'],other_fields['PF_TOTAL'])
other_fields['AM_required']=np.where(other_fields['ACTUALNEED']<other_fields['AM_TOTAL'],other_fields['ACTUALNEED'],other_fields['AM_TOTAL'])
```
